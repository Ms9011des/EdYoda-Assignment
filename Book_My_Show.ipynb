```python
class BookTicket:
    def __init__(self):
        self.r = int(input("give number of rows: "))
        self.c = int(input("give numbers of columns: "))
        while self.r < 1 or self.c < 1:
            print("Give proper input")
            self.r = int(input("give number of rows: "))
            self.c = int(input("give numbers of columns: "))
        self.seats = {}
        if self.r * self.c < 61:
            for i in range(1, self.r + 1):
                for j in range(1, self.c + 1):
                    self.seats['{},{}'.format(i, j)] = ["S", 10]
        elif self.r * self.c > 60:
            full_price_seats = self.r//2
            for i in range(1, full_price_seats+1):
                for j in range(1, self.c + 1):
                    self.seats['{},{}'.format(i, j)] = ["S", 10]
            for i in range(self.r-full_price_seats+1, self.r +1):
                for j in range(1, self.c + 1):
                    self.seats['{},{}'.format(i, j)] = ["S", 8]

    def show(self):
        n = 1
        print(" ", end=" ")
        for i in range(1, self.c + 1):
            print(i, end=" ")

        print("\r")
        for i in range(1, self.r + 1):
            print(n, end=" ")
            for j in range(1, self.c + 1):
                print(self.seats["{},{}".format(i, j)][0], end=" ")
            print("\r")
            n +=1

    def book(self):
        selection_row = input("Select the row: ")
        selection_column = input("Select the column: ")
        if self.seats['{},{}'.format(selection_row, selection_column)][0] == "B":
            print("Seat is already booked.")
        else:
            customer_name = input("Enter your name: ")
            customer_age = input("Enter your age: ")
            customer_gender = input("Enter your gender: ")
            customer_contact_no = input("Enter your contact Number: ")
            print("")
            print("Name: {}\nAge: {}\nGender: {}\nContact Number: {}\nSEAT: Row {}, Column {}".format(customer_name,
                                                                                                      customer_age,
                                                                                                      customer_gender,
                                                                                                      customer_contact_no,
                                                                                                      selection_row,
                                                                                                      selection_column))
            print("")
            confirmation = input("Please confirm above info by entering 1/ enter anything else to cancel:")
            if confirmation == "1":
                customer_info = [customer_name, customer_age, customer_gender, customer_contact_no]
                self.seats["{},{}".format(selection_row, selection_column)].extend(customer_info)
                self.seats["{},{}".format(selection_row, selection_column)][0] = 'B'
                print("Ticket Booked")
            else:
                print("Ticket Canceled")

    def statistics(self):
        booked_tickets = 0
        current_income = 0
        total_income = 0
        for i in self.seats.values():
            total_income += i[1]
            if i[0] == "B":
                booked_tickets += 1
                current_income += i[1]
        booking_percentage = round((booked_tickets / (self.r * self.c)) * 100, 2)
        print('Number of purchased tickets: ', booked_tickets)
        print('Percentage: ', booking_percentage, '%')
        print('Current income: ', current_income, '$')
        print('Total income: ', total_income, '$')

    def user_info(self):
        row = input("give row number: ")
        column = input("give column number: ")
        user_info = self.seats['{},{}'.format(row, column)]
        if user_info[0] == "B":
            print('Name: {}'.format(user_info[2]))
            print('Gender: {}'.format(user_info[4]))
            print('Age: {}'.format(user_info[3]))
            print('Ticket Price: {}'.format(user_info[1]))
            print('Phone No: {}'.format(user_info[5]))
        else:
            print("Seat is not booked yet.")


a = BookTicket()
options = ["Show the seats", "Buy a ticket", "statistics", "Show Booked ticket user Info"]
for number, value in enumerate(options):
    print(str(number + 1) + ":\t" + value)
print("0" + ':\t' + "Exit")
value = input("Please select from above: ")
print("")
while True:
    if value == '1':
        a.show()
        print("")
    elif value == '2':
        a.book()
        print("")
    elif value == '3':
        a.statistics()
        print("")
    elif value == '4':
        a.user_info()
        print("")
    elif value == '0':
        break

    for number, value in enumerate(options):
        print(str(number + 1) + ":\t" + value)
    print('0:\tExit')
    value = input("Please select from above: ")
    print("")
```
